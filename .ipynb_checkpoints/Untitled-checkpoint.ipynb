Using the `update_vals` function below, 'Contributing Factors. Dissatisfaction' and 'Contributing Factors. Job Dissatisfaction' in the tafe_resignations dataframe so that each contains only True, False, or NaN values.

```python
def update_vals(val):
    if pd.isnull(val):
        return np.nan
    elif val == '-':
        return False
    else:
        return True
```

```python
# Next we apply the function to the columns we want and store it in a new variable
tafe_cols = ['Contributing Factors. Dissatisfaction','Contributing Factors. Job Dissatisfaction']
tafe_resignations[tafe_cols] = tafe_resignations[tafe_cols].applymap(update_vals)
```

```python
tafe_resignations[tafe_cols]
```

```python
tafe_resignations['dissatisfied'] = tafe_resignations[tafe_cols].copy().any(axis=1, skipna=False).copy()
```

Using this syntax, `df.any(axis=1, skipna=False)`, after our changes, the new dissatisfied column will contain just the following values for tafe dataset:

* `True`: indicates a person resigned because they were dissatisfied with the job
* `False`: indicates a person resigned because of a reason other than dissatisfaction with the job
* `NaN`: indicates the value is missing

```python
#Refer: https://stackoverflow.com/questions/54185221/extracting-only-numbers-from-series-python
combined_updated['institute_service_new']=combined_updated['institute_service'].copy().astype(str).str.extract(r'(\d+)')
```

To analyze the data, we'll convert these numbers into categories. We'll base our anlaysis on this [article](https://www.businesswire.com/news/home/20171108006002/en/Age-Number-Engage-Employees-Career-Stage), which makes the argument that understanding employee's needs according to career stage instead of age is more effective.

We'll use the slightly modified definitions below:

* New: Less than 3 years at a company
* Experienced: 3-6 years at a company
* Established: 7-10 years at a company
* Veteran: 11 or more years at a company

```python
def career_stage(val):
    if val < 3:
        return "New: Less than 3 years at a company"
    elif (val >= 3 and val < 7):
        return 'Experienced: 3-6 years at a company'
    elif (val >= 7 and val < 11):
        return 'Established: 7-10 years at a company'
    elif val > 11:
        return 'Veteran: 11 or more years at a company'
    else: #pd.isnull(val):
        return np.nan
```

```python
combined_updated['service_cat'] = combined_updated['institute_service_new'].copy().apply(career_stage)
```

```python
combined_updated[['institute_service_new','service_cat']]
```

Recall that the dissatisfied column consists of Boolean values, meaning they're either True or False. Methods such as the `df.pivot_table()` method actually treat Boolean values as integers, so a **True** value is considered to be **1** and a **False** value is considered to be **0**. That means that we can aggregate the dissatisfied column and calculate the number of people in each group, the percentage of people in each group, etc.

```python
combined_updated['dissatisfied'].value_counts(dropna = False)
```

Use the `DataFrame.fillna()` method to replace the missing values in the dissatisfied column with the value that occurs most frequently. In this case **False** is the most occuring

```python
combined_updated['dissatisfied'] = combined_updated['dissatisfied'].copy().fillna(False)
```

```python
combined_updated['dissatisfied'].value_counts(dropna = False)
```

Use the `DataFrame.pivot_table()` method to calculate the percentage of dissatisfied employees in each service_cat group.

Since a **True** value is considered to be **1**, calculating the mean will also calculate the percentage of dissatisfied employees. The default aggregation function is the mean, so we can exclude the aggfunc argument.

```python
combined_pivot_table = combined_updated.pivot_table(values = 'dissatisfied', index = 'service_cat')
```

```python
combined_pivot_table
```

```python
%matplotlib inline 
combined_pivot_table.plot(kind = 'barh')
```

To analyze the data for age, we'll convert these numbers into categories. We'll base our anlaysis on this [website](https://www.ons.gov.uk/employmentandlabourmarket/peopleinwork/employmentandemployeetypes/bulletins/employmentintheuk/march2019), which gives us the UK employment rate by age band.

We'll use the slightly modified definitions below:

* Ages 20 years and younger to 24 years old at a company
* Ages 25 to 35 years old at a company
* Ages 35 years to 49 years old at a company
* Ages 50 years to 64 years old at a company

```python
def age_band(age):
    if age <= 20 and age < 25:
        return "Ages 20 years and younger to 24 years old at a company"
    elif (age >= 25 and age < 36):
        return 'Ages 25 to 35 years old at a company'
    elif (age >= 35 and age < 50):
        return 'Ages 35 years to 49 years old at a company'
    elif (age >= 50 and age < 65):
        return 'Ages 50 years to 64 years old at a company'
    else: #pd.isnull(age):
        return np.nan
```

```python
combined_updated['age_group'] = combined_updated['age_updated'].copy().apply(age_band)
```

```python
age_pivot_table = combined_updated.pivot_table(values = 'dissatisfied', index = 'age_group')
```
